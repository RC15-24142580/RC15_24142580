```python
import pandas as pd
import plotly.subplots as sp
import plotly.graph_objects as go

# ---------------------------------------------------------
# 1. Data Loading & Configuration
# ---------------------------------------------------------
filename = 'final_002.csv'
df = pd.read_csv(filename)

# Settings for time-series conversion
sampling_interval = 8.25 # seconds per sample
metrics = ["Void", "Bio", "Tech", "Entropy", "GSR"]
colors = ["#636EFA", "#EF553B", "#00CC96", "#AB63FA", "#FFA15A"]

# Calculate time axis in seconds
x_seconds = [i * sampling_interval for i in range(len(df))]
total_seconds = len(df) * sampling_interval

# Generate ticks every 60 seconds (1 minute intervals)
tick_vals = [i for i in range(0, int(total_seconds) + 1, 60)]
tick_text = [f"{i//60}min" for i in tick_vals]

# ---------------------------------------------------------
# 2. Subplot Creation
# ---------------------------------------------------------
# Create stacked subplots for each metric with shared X-axis
fig = sp.make_subplots(
    rows=len(metrics), 
    cols=1, 
    shared_xaxes=True, 
    vertical_spacing=0.04, 
    subplot_titles=metrics
)

for i, metric in enumerate(metrics):
    fig.add_trace(
        go.Scatter(
            x=x_seconds, 
            y=df[metric], 
            name=metric, 
            mode='lines', 
            line=dict(color=colors[i], width=1)
        ), 
        row=i+1, col=1
    )

# ---------------------------------------------------------
# 3. Layout & Export Configuration
# ---------------------------------------------------------
fig.update_layout(
    width=1600, height=900,
    title_text=f"Multi-Metric Time-Series Analysis: {filename}",
    template="plotly_dark", 
    paper_bgcolor="black", 
    plot_bgcolor="black", 
    showlegend=False
)

# Set common X-axis properties
fig.update_xaxes(tickvals=tick_vals, ticktext=tick_text, gridcolor='#333333')
fig.update_yaxes(gridcolor='#333333')

# --- Key Feature: SVG Export Configuration ---
# Setting the download button to export high-quality SVG instead of PNG
config = {
  'toImageButtonOptions': {
    'format': 'svg',
    'filename': 'multi_metric_export',
    'height': 900,
    'width': 1600,
    'scale': 1 
  }
}

# Render plot with custom config
fig.show(config=config)
```

```python
import pandas as pd
import plotly.graph_objects as go

# ---------------------------------------------------------
# 1. Data Loading & Time Axis Configuration
# ---------------------------------------------------------
filename = 'final_004.csv'
df = pd.read_csv(filename)

# Settings for time-series conversion
sampling_interval = 8 # seconds per sample
metrics = ["Void", "Bio", "Tech", "Entropy", "GSR"]
colors = ["#636EFA", "#EF553B", "#00CC96", "#AB63FA", "#FFA15A"]

# Calculate time axis in seconds
x_seconds = [i * sampling_interval for i in range(len(df))]
total_seconds = len(df) * sampling_interval

# Generate minute-based ticks for better readability
tick_vals = [i for i in range(0, int(total_seconds) + 1, 60)]
tick_text = [f"{i//60}min" for i in tick_vals]

# ---------------------------------------------------------
# 2. Visualization (Normalized Overlaid Plot)
# ---------------------------------------------------------
fig = go.Figure()

for i, metric in enumerate(metrics):
    # Min-Max Normalization to bring all metrics to the same scale [0, 1]
    min_val, max_val = df[metric].min(), df[metric].max()
    norm_y = (df[metric] - min_val) / (max_val - min_val) if (max_val - min_val) != 0 else 0
    
    # Emphasize GSR with a thicker line
    line_width = 1.5 if metric == "GSR" else 1
    
    fig.add_trace(go.Scatter(
        x=x_seconds, 
        y=norm_y, 
        name=metric, 
        mode='lines', 
        line=dict(color=colors[i], width=line_width)
    ))

# ---------------------------------------------------------
# 3. Layout & UI Configuration
# ---------------------------------------------------------
fig.update_layout(
    height=600, 
    title_text=f"Overlaid Normalized Analysis: {filename}",
    template="plotly_dark", 
    paper_bgcolor="black", 
    plot_bgcolor="black",
    # Horizontal legend for better spatial efficiency
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)

fig.update_xaxes(tickvals=tick_vals, ticktext=tick_text, gridcolor='#333333')
fig.update_yaxes(gridcolor='#333333', title="Normalized Intensity (0-1)")

# --- Export Settings: High-Quality SVG ---
config = {
  'toImageButtonOptions': {
    'format': 'svg',
    'filename': 'overlaid_export',
    'height': 900,
    'width': 1600,
    'scale': 1 
  }
}

fig.show(config=config)
```
